1.Calcula las siguientes estadísticas descriptivas para las variables numéricas:

Media

Mediana

Desviación estándar

Mínimo

Máximo

```python
from itertools import combinations

# definimos la lista los nombres de las variables numéricas 
numericas = df.select_dtypes(include = np.number).columns
numericas

```

```python
for columna in numericas:
   
    print(f"la media de la {columna} es {df[columna].mean().round(2)}")
    print(f"la mediana de la {columna} es {df[columna].median()}")
    print(f"la desviación estandar de la {columna} es {df[columna].std()}")
    print(f"el min de la {columna} es {df[columna].min()}")
    print(f"el max de la {columna} es {df[columna].max()}")
    print ("----------------------------------------------------------")
```

```python
# Suponiendo que 'numericas' contiene las columnas numéricas de tu DataFrame
for columna in numericas:
    print(f"\n📊 ESTADÍSTICAS PARA: '{columna.upper()}'")
    print("-" * 40)
    print(f"Media → {df[columna].mean():.2f}")
    print(f"Mediana → {df[columna].median():.2f}")
    print(f"Desviación estándar → {df[columna].std():.2f}")
    print(f"Mínimo → {df[columna].min():.2f}")
    print(f"Máximo → {df[columna].max():.2f}")
    print(f"Rango intercuartílico → {df[columna].quantile(0.75) - df[columna].quantile(0.25):.2f}")
```

4.Realiza un análisis de la mortalidad infantil (infant_mortality) y clasifica los países en tres categorías: baja, moderada y alta mortalidad infantil.

```python
df["infant_mortality"].describe()
```

```python

def mortalidad_infantil(infant_mortality):
    if infant_mortality < 14:
        return 'baja'
    elif infant_mortality < 32:
        return 'moderada'
    else:
        return 'alta'
df['categoria_mortalidad'] = df['infant_mortality'].apply(mortalidad_infantil)
display(df)
```
