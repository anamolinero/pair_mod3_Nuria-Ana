```python
from itertools import combinations

# definimos la lista los nombres de las variables num√©ricas 
numericas = df.select_dtypes(include = np.number).columns
numericas

```

```python
for columna in numericas:
   
    print(f"la media de la {columna} es {df[columna].mean().round(2)}")
    print(f"la mediana de la {columna} es {df[columna].median()}")
    print(f"la desviaci√≥n estandar de la {columna} es {df[columna].std()}")
    print(f"el min de la {columna} es {df[columna].min()}")
    print(f"el max de la {columna} es {df[columna].max()}")
    print ("----------------------------------------------------------")
```

```python
# Suponiendo que 'numericas' contiene las columnas num√©ricas de tu DataFrame
for columna in numericas:
    print(f"\nüìä ESTAD√çSTICAS PARA: '{columna.upper()}'")
    print("-" * 40)
    print(f"Media ‚Üí {df[columna].mean():.2f}")
    print(f"Mediana ‚Üí {df[columna].median():.2f}")
    print(f"Desviaci√≥n est√°ndar ‚Üí {df[columna].std():.2f}")
    print(f"M√≠nimo ‚Üí {df[columna].min():.2f}")
    print(f"M√°ximo ‚Üí {df[columna].max():.2f}")
    print(f"Rango intercuart√≠lico ‚Üí {df[columna].quantile(0.75) - df[columna].quantile(0.25):.2f}")
```

4.Realiza un an√°lisis de la mortalidad infantil (infant_mortality) y clasifica los pa√≠ses en tres categor√≠as: baja, moderada y alta mortalidad infantil.

```python
df["infant_mortality"].describe()
```

```python

def mortalidad_infantil(infant_mortality):
    if infant_mortality < 14:
        return 'baja'
    elif infant_mortality < 32:
        return 'moderada'
    else:
        return 'alta'
df['categoria_mortalidad'] = df['infant_mortality'].apply(mortalidad_infantil)
display(df)
```
