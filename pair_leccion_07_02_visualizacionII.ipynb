```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

1. ¿Cuál es el continente con la mayor cantidad de países en el conjunto de datos? Proporciona el porcentaje de países en ese continente.

```python
df = pd.read_csv("archivos/world_data_full_apply_nulos.csv",index_col=0) 
df.head()
```

```python
paises_continentes = df.groupby('continent')['country'].count().reset_index()
paises_continentes
```

```python
#PIEPLOT: con este gráfico hace automáticamente el porcentaje, no tenemos que calcularlo nosotras
sns.set_palette("pastel")
plt.figure(figsize=(8, 8))
plt.pie(
    paises_continentes["country"],
    labels=paises_continentes["continent"],
    autopct='%1.1f%%',
    textprops={'fontsize': 8},
    startangle=90
)
plt.title('Distribución de países por continente')
plt.show()
```

2. ¿Cuáles son los 10 países con la mayor área forestal?

```python
top_forestal = df.sort_values('forested_area', ascending=False).head(10)
top_forestal
```

```python
sns.barplot(x = "forested_area", 
            y = "country", 
            data = top_forestal,  
            palette = "magma")
plt.title('Top 10 países con mayor área forestal')
```

3. ¿Cómo se distribuye la densidad de población entre los países?

```python
sns.histplot(x = "density", data = df, color = "green",bins=20)
plt.xlabel("Densidad")
plt.title("Distribución densidad de población por país")
```
