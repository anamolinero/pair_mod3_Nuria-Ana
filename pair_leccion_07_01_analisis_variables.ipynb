```python
# importamos las librerías que necesitamos

# Tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd

# Visualización
# ------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
```

```python
df = pd.read_csv("world_data_full_apply_nulos.csv",index_col=0) 
df.head(3)
```

¿Cómo se distribuye la tasa de natalidad en el conjunto de datos? ¿Hay algún patrón evidente?

```python
sns.histplot(x = "birth_rate", data = df, color = "red",bins=20)
plt.xlabel("tasa de natalidad")
plt.ylabel("cantidad")
plt.title("distribucion tasa de natalidad");
```

¿Existe una relación aparente entre la tasa de mortalidad infantil y la esperanza de vida? Proporciona una breve descripción de los patrones que observes.

```python
plt.scatter(x = "infant_mortality", 
            y = "life_expectancy", 
            data = df, 
            c = "turquoise", 
            s = 2, 
            marker = "^" )

# ponemos los nombres de los ejes
plt.xlabel("mortalidad infantil")
plt.ylabel("esperanza de vida en años")

# ponemos un título usando el método 'plt.title()'
# en este caso lo que hemos añadido es al parámetro 'fontsize' que nos permite cambiar el tamaño del título
plt.title("relación entre mortalidad infantil y esperanza de vida", fontsize = 8);
```

¿Cuál es la mediana de la esperanza de vida en Europa? Proporciona un rango intercuartil aproximado y verifica si existen valores atípicos.

```python
europa=df[df["continent"]=="Europe"]#creamos con los paises europeos
europa.head(2)
```

```python


sns.boxplot(x="life_expectancy", data=europa, color="pink")

plt.xlabel("esperanza de vida en años")
plt.title("mediana esperanza de vida", fontsize = 8);
            


```
